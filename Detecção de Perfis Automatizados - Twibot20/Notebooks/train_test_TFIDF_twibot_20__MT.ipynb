## TFIDF

The intuition behind this measure is : If a word appears frequently in a document, then it should be important and we should give that word a high score. But if a word appears in too many other documents, itâ€™s probably not a unique identifier, therefore we should assign a lower score to that word.

TF(w) = (Number of times term w appears in a document) / (Total number of terms in the document)


IDF(w) = log_e(Total number of documents / Number of documents with term w in it)

**TFIDF(w) = TF(w) * IDF(w)**

https://www.geeksforgeeks.org/understanding-tf-idf-term-frequency-inverse-document-frequency/

https://countwordsfree.com/stopwords

````python
train_twt = []
for twt in df_train.tweet:
  if str(twt) == "nan":
    train_twt.append("NULO")
  else:
    train_twt.append(twt)
````

````python
len(train_twt), len(df_train)
````

````python
vectorizer = TfidfVectorizer(strip_accents = "ascii", max_features=200, stop_words='english')
tfidf= vectorizer.fit(train_twt)
````

````python
train_twt = tfidf.transform(train_twt)
#numpy
dense = train_twt.todense()
#pandas
TFIDF_TRAIN = pd.DataFrame(dense,columns = vectorizer.get_feature_names_out())

tfidf_train = TFIDF_TRAIN.copy()
tfidf_train.head(2)
````

````python
tfidf_train.describe().sort_values(by="mean",axis=1,ascending=False)
````

````python
tfidf_train.head(2)
````

````python
#TESTE
test_twt = []
for twt in df_test.tweet:
  if str(twt) == "nan":
    test_twt.append("NULO")
  else:
    test_twt.append(twt)

test_twt = tfidf.transform(test_twt)

#numpy
dense = test_twt.todense()
#pandas
TFIDF_TEST = pd.DataFrame(dense,columns = tfidf.get_feature_names_out())
tfidf_test = TFIDF_TEST.copy()
````

##Variance Feature Select
A ideia Ã© retirar as colunas com uma variancia muito baixa.

````python
tfidf_train = TFIDF_TRAIN.copy()
tfidf_test =  TFIDF_TEST.copy()
````

````python
from sklearn.feature_selection import VarianceThreshold
var_select = VarianceThreshold(.001602) # removendo Feautures com variancia menor que o 3Âº quartil
````

````python
var_select.fit(tfidf_train)
mask=var_select.get_support()
````
